```python
import pandas as pd

path ="/data/StudentPerformanceFactors.csv"
df = pd.read_csv(path)
df.head(10)
```

Kode ini untuk menampilkan columns yang spesifik bila ingin melihat apa saja column maka gunakan df.columns()

```python
tutoring_Sessions= df["Tutoring_Sessions"] # mengambil data yang ada pada kolom Tutoring_Sessions
parental_education_level = df['Parental_Education_Level'] # mengambil data yang ada pada kolom Parental_Education_Level
gender_studens = df['Gender'] # mengambil data yang ada pada kolom Gender
```

**Untuk mengambil data tutoring sessions**

```python
panjang_x = len(tutoring_Sessions)
value_tutoring = tutoring_Sessions.values # untuk mengambil isi kolom pada table tutoring_sesion
```

**Untuk Mengambil data perental education level**

```python
 # untuk mengambil data dari kolom parental_education_level
 # .drop_duplicate agar tidak ada yang sama
parental_level_label = ['High School', 'College', 'Postgraduate']
# parental_level_label = parental_education_level.drop_duplicates().values
# Menghitung jumlah total data
panjang_index_perental = len(parental_education_level)

# mengambil banyaknya jumlah parent yang high school, college, postgraduate
parental_high_school = parental_education_level[parental_education_level == "High School"].count()
parental_college = parental_education_level[parental_education_level == "College"].count()
parental_postgraduate = parental_education_level[parental_education_level == "Postgraduate"].count()

# Menghitung jumlah NaN (jika ada)
parental_nan = parental_education_level[parental_education_level.isna()].count()

value_parent_level = [parental_high_school, parental_college, parental_postgraduate]
```

**Untuk mengambil banyak data gender**

```python
lebel_gender = ['Male', 'Female']
panjang_index_gender = len(gender_studens)
gender_male = gender_studens[gender_studens == "Male"].count()
gender_female = gender_studens[gender_studens == "Female"].count()
velue_gender = [gender_male, gender_female]
```

**Diagram batang**

diagram untuk melihat siswa yang mengikuti tutoring session atau tidak


```python
# Membuat grafik garis
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
plt.plot(range(0, panjang_x), value_tutoring, linestyle='-', color='blue', label="Sessions")  # Garis dengan marker
plt.title("Grafik Tutoring Sessions")
plt.xlabel("Siswa ke-(1,2,3,...,n)")
plt.ylabel("Jumlah Sesi yang di ikuti")
plt.legend()
plt.grid(True)
plt.show()
```

**Pie Charm**
untuk melihat persentase tingkat edukasi/jenjang orang tua siswa/siswi/mahasiswa

```python
plt.figure(figsize=(8, 8))
plt.pie(value_parent_level, labels=parental_level_label, colors=['red', 'blue', 'green'], autopct='%1.1f%%', startangle=90)

# Menambahkan judul
plt.title("Parental Education Level")
plt.legend()
# Menampilkan pie chart
plt.show()
```

**Diagram Garis**

```python

# Membuat line chart
plt.figure(figsize=(8, 6))
plt.plot(parental_level_label, value_parent_level, color='blue', marker='o', linestyle='-', linewidth=2)

# Menambahkan judul
plt.title("Parental Education Level", fontsize=14)

# Menambahkan label pada sumbu
plt.xlabel("Parental Education Level", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Menambahkan grid
plt.grid(True, linestyle='--', alpha=0.5)

# Menampilkan line chart
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.bar(lebel_gender, velue_gender,  color=['blue', 'pink'], label= lebel_gender)
plt.title("Grafik Tutoring Sessions")
plt.xlabel("Gender")
plt.ylabel("Banyak")
plt.legend()
plt.grid(True)
plt.show()
```
